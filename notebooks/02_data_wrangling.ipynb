In this notebook, I'll clean and prepare the survey data for analysis. Data wrangling is an important step where I fix inconsistencies, handle missing values, and get the data ready for deeper analysis.

## Importing Libraries

First, I'll import the libraries I need for data cleaning and manipulation.

```python
import pandas as pd
import numpy as np
```

## Loading the Dataset

Let me load the survey data that I explored in the previous notebook.

```python
# Load the survey data
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv')

# Check the shape
print(f"Dataset shape: {df.shape}")
df.head()
```

## Finding Duplicate Rows

First, I'll check if there are any duplicate rows in the dataset. Duplicates can skew our analysis.

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

## Removing Duplicates

If there are duplicates, I'll remove them to keep only unique responses.

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Verify duplicates are removed
print(f"Rows after removing duplicates: {df.shape[0]}")
```

## Finding Missing Values

Now let me check which columns have missing values and how many.

```python
# Count missing values in each column
missing_values = df.isnull().sum()

# Show only columns with missing values
missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing Count': missing_values.values,
    'Percentage': (missing_values.values / len(df) * 100).round(2)
})

missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)
print(missing_df.head(10))
```

## Checking Data Types

Let me see what data types we have. Some columns might need to be converted to the right type.

```python
# Check data types
print(df.dtypes.value_counts())
print("\nSample of data types:")
print(df.dtypes.head(20))
```

## Cleaning the 'EdLevel' Column

I'll look at the education level column and see if there are any inconsistencies.

```python
# Check unique values in EdLevel
print("Education levels in the data:")
print(df['EdLevel'].value_counts())
```

## Cleaning the 'Country' Column

Let me check how many different countries are represented.

```python
# Check number of unique countries
print(f"Number of unique countries: {df['Country'].nunique()}")
print("\nTop 10 countries:")
print(df['Country'].value_counts().head(10))
```

## Handling Missing Values in 'Employment'

For the Employment column, I'll check how many missing values exist.

```python
# Check missing values in Employment
employment_missing = df['Employment'].isnull().sum()
print(f"Missing values in Employment: {employment_missing}")

# Show what values exist
print("\nEmployment categories:")
print(df['Employment'].value_counts())
```

## Standardizing Column Names

I'll make all column names lowercase and replace spaces with underscores for easier coding.

```python
# Standardize column names
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Show first few column names
print("Column names after standardization:")
print(df.columns[:10].tolist())
```

## Creating a Cleaned Dataset

Now I'll create a cleaned version focusing on the most important columns for my analysis.

```python
# Select key columns for analysis
key_columns = ['respondent', 'mainbranch', 'age', 'country', 'edlevel', 
               'employment', 'yearscodepro', 'languageworkedwith', 
               'databaseworkedwith', 'platformworkedwith']

# Create cleaned dataframe with available columns
available_columns = [col for col in key_columns if col in df.columns]
df_clean = df[available_columns].copy()

print(f"Cleaned dataset shape: {df_clean.shape}")
df_clean.head()
```

## Summary of Missing Values in Clean Dataset

Let me check the missing values in my cleaned dataset.

```python
# Summary of missing values
clean_missing = df_clean.isnull().sum()
clean_missing_pct = (clean_missing / len(df_clean) * 100).round(2)

missing_summary = pd.DataFrame({
    'Missing Count': clean_missing,
    'Percentage': clean_missing_pct
})

print(missing_summary[missing_summary['Missing Count'] > 0])
```

## Saving the Cleaned Dataset

Finally, I'll save the cleaned dataset for use in future analysis.

```python
# Save cleaned dataset
df_clean.to_csv('survey_data_cleaned.csv', index=False)
print("✅ Cleaned dataset saved as 'survey_data_cleaned.csv'")
```

In this notebook, I completed several data cleaning tasks:

- Checked for and removed duplicate rows
- Identified missing values across all columns
- Examined data types to ensure consistency
- Standardized column names for easier coding
- Created a cleaned dataset with key columns
- Saved the cleaned data for future analysis

The data is now ready for more detailed analysis and visualization!
