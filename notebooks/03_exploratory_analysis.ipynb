## Importing Libraries

I'll start by importing the necessary libraries for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for better looking plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## Loading the Dataset

Now I'll load the cleaned survey dataset and take a look at its structure.

```python
# Load the survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display all columns
pd.set_option('display.max_columns', None)

# Show first few rows
print(f"Dataset shape: {df.shape}")
df.head()
```

## Understanding the Data Structure

Let me check the basic information about this dataset.

```python
# Get dataset info
print("Dataset Information:")
print(df.info())
print("\nColumn Names:")
print(df.columns.tolist())
```

## Handling Missing Data

I'll identify missing values in key columns and decide how to handle them.

```python
# Check for missing values
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df) * 100).round(2)

missing_summary = pd.DataFrame({
    'Missing Count': missing_data,
    'Percentage': missing_percent
})

# Show columns with missing values
print("Columns with missing values:")
print(missing_summary[missing_summary['Missing Count'] > 0].sort_values('Missing Count', ascending=False))
```

Now I'll handle missing values in important columns. For columns with few missing values, I'll drop those rows. For others, I'll keep them as they might still provide useful information.

```python
# Create a copy for analysis
df_clean = df.copy()

# Drop rows where Employment is missing (if this column exists)
if 'Employment' in df_clean.columns:
    df_clean = df_clean.dropna(subset=['Employment'])

print(f"Rows after handling missing values: {df_clean.shape[0]}")
```

## Experience and Job Satisfaction Analysis

I'll analyze the relationship between years of coding experience and job satisfaction.

```python
# Check if YearsCodePro column exists
if 'YearsCodePro' in df_clean.columns:
    # Remove missing values for this analysis
    exp_data = df_clean[df_clean['YearsCodePro'].notna()].copy()
    
    # Convert to numeric if needed
    exp_data['YearsCodePro'] = pd.to_numeric(exp_data['YearsCodePro'], errors='coerce')
    
    # Create experience ranges
    bins = [0, 5, 10, 20, 100]
    labels = ['0-5 years', '5-10 years', '10-20 years', '20+ years']
    exp_data['ExperienceRange'] = pd.cut(exp_data['YearsCodePro'], bins=bins, labels=labels)
    
    # Count by experience range
    exp_distribution = exp_data['ExperienceRange'].value_counts().sort_index()
    print("Experience Distribution:")
    print(exp_distribution)
else:
    print("YearsCodePro column not found in dataset")
```

Let me visualize the experience distribution.

```python
# Visualize experience distribution
if 'YearsCodePro' in df_clean.columns:
    plt.figure(figsize=(10, 6))
    exp_distribution.plot(kind='bar', color='steelblue')
    plt.title('Distribution of Professional Coding Experience')
    plt.xlabel('Experience Range')
    plt.ylabel('Number of Respondents')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## Visualizing Job Satisfaction

Now I'll look at the distribution of job satisfaction levels.

```python
# Check if JobSat column exists
if 'JobSat' in df_clean.columns:
    # Get job satisfaction distribution
    jobsat_dist = df_clean['JobSat'].value_counts()
    print("Job Satisfaction Distribution:")
    print(jobsat_dist)
    
    # Create visualization
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_clean, x='JobSat', hue='JobSat', palette='viridis', legend=False)
    plt.title('Distribution of Job Satisfaction Levels')
    plt.xlabel('Job Satisfaction')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("JobSat column not found in dataset")
```

## Analyzing Remote Work Preferences

I'll explore remote work trends and how they relate to employment types.

```python
# Check if RemoteWork column exists
if 'RemoteWork' in df_clean.columns:
    remote_dist = df_clean['RemoteWork'].value_counts()
    print("Remote Work Distribution:")
    print(remote_dist)
    
    # Visualize
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_clean, x='RemoteWork', hue='RemoteWork', palette='Set2', legend=False)
    plt.title('Remote Work Preferences')
    plt.xlabel('Remote Work Status')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("RemoteWork column not found in dataset")
```

Let me check the relationship between employment type and remote work preferences.

```python
# Cross-tabulation of Employment and RemoteWork
if 'Employment' in df_clean.columns and 'RemoteWork' in df_clean.columns:
    crosstab = pd.crosstab(df_clean['Employment'], df_clean['RemoteWork'])
    print("\nEmployment vs Remote Work:")
    print(crosstab)
else:
    print("Required columns not found for cross-tabulation")
```

## Programming Language Analysis

I'll analyze which programming languages are most popular among respondents.

```python
# Check for language columns
lang_col = None
for col in ['LanguageHaveWorkedWith', 'LanguageWorkedWith', 'LanguageWantToWorkWith']:
    if col in df_clean.columns:
        lang_col = col
        break

if lang_col:
    # Get all languages (split by delimiter if multiple)
    languages = df_clean[lang_col].dropna()
    print(f"\nAnalyzing column: {lang_col}")
    print(f"Sample values:\n{languages.head()}")
    
    # If languages are separated by semicolon, split and count
    if languages.str.contains(';').any():
        all_langs = languages.str.split(';').explode()
        lang_counts = all_langs.value_counts().head(10)
        
        print("\nTop 10 Programming Languages:")
        print(lang_counts)
        
        # Visualize
        plt.figure(figsize=(12, 6))
        lang_counts.plot(kind='bar', color='coral')
        plt.title('Top 10 Most Popular Programming Languages')
        plt.xlabel('Programming Language')
        plt.ylabel('Number of Users')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        lang_counts = languages.value_counts().head(10)
        print("\nTop 10 Programming Languages:")
        print(lang_counts)
else:
    print("No language column found in dataset")
```

## Country-wise Analysis

Let me see which countries have the most respondents.

```python
# Analyze by country
if 'Country' in df_clean.columns:
    country_dist = df_clean['Country'].value_counts().head(10)
    print("Top 10 Countries by Respondents:")
    print(country_dist)
    
    # Visualize
    plt.figure(figsize=(12, 6))
    country_dist.plot(kind='barh', color='teal')
    plt.title('Top 10 Countries by Number of Respondents')
    plt.xlabel('Number of Respondents')
    plt.ylabel('Country')
    plt.tight_layout()
    plt.show()
else:
    print("Country column not found in dataset")
```

## Education Level Analysis

I'll explore the relationship between education level and employment type.

```python
# Analyze education level
if 'EdLevel' in df_clean.columns:
    ed_dist = df_clean['EdLevel'].value_counts()
    print("Education Level Distribution:")
    print(ed_dist)
    
    # Cross-tab with Employment if available
    if 'Employment' in df_clean.columns:
        ed_emp_crosstab = pd.crosstab(df_clean['EdLevel'], df_clean['Employment'])
        print("\nEducation Level vs Employment Type:")
        print(ed_emp_crosstab)
else:
    print("EdLevel column not found in dataset")
```

## Age Distribution

Let me look at the age distribution of respondents.

```python
# Analyze age distribution
if 'Age' in df_clean.columns:
    age_dist = df_clean['Age'].value_counts().sort_index()
    print("Age Distribution:")
    print(age_dist)
    
    # Visualize
    plt.figure(figsize=(12, 6))
    age_dist.plot(kind='bar', color='mediumpurple')
    plt.title('Age Distribution of Survey Respondents')
    plt.xlabel('Age Group')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Age column not found in dataset")
```

## Saving the Analyzed Dataset

Finally, I'll save the cleaned dataset for future use.

```python
# Save the cleaned dataset
df_clean.to_csv('survey_data_analyzed.csv', index=False)
print("✅ Analyzed dataset saved as 'survey_data_analyzed.csv'")
print(f"Final dataset shape: {df_clean.shape}")
```
